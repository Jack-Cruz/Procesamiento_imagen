```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(im, size = 12):
  _, ax = plt.subplots(figsize=(size, size))
  ax.imshow(im, interpolation = "nearest")
  plt.tight_layout()
  plt.show()
```

```python
def loadImg(img):
  im = cv2.imread(img)
  return im[:,:,::-1]
```

```python
base = "https://raw.githubusercontent.com/lmcanavals/dip/main/imagenes/"
```

```python
# Cargar imágen de github a google colab
img1 = "gradient.png"
url = f"{base}{img1}"
!curl -O $url
```

```python
gradient = loadImg(img1)
show(gradient, 4)
```

```python
# Cargar imágen de github a google colab
img2 = "eminem.jpg"
url = f"{base}{img2}"
!curl -O $url
```

```python
eminem = loadImg(img2)
show(eminem, 4)
```

```python
img3 = "zuckerberg.jpg"
url = f"{base}{img3}"
!curl -O $url
```

```python
zuckerberg = loadImg(img3)
show(zuckerberg, 4)
```

# Image Operations

## Adition

```python
eminem = eminem / 255
gradient = gradient / 255
zuckerberg = zuckerberg / 255
```

```python
fig = plt.figure(figsize = (12, 10))

for i in range(12):
  alfa = (i + 0.5) / 12
  a = fig.add_subplot(3, 4, i + 1)
  
  aux = alfa * gradient + (1 - alfa) * eminem

  # Inverso 
  # aux = 255 - aux

  plt.imshow(aux)
  plt.text(0, 0, f"Alfa (a) = {alfa:.3f}")
```

## Adition 2

```python
aux = (zuckerberg + eminem)  / 2
plt.imshow(aux)
```

```python
aux = (zuckerberg + eminem)  / 2
plt.imshow(aux)
```

## Difference

```python
aux = abs(zuckerberg - eminem)
plt.imshow(aux)
```

```python
fig = plt.figure(figsize=(12, 6))

a = fig.add_subplot(1, 2, 1)
plt.axis("off")

aux = abs(eminem - zuckerberg)
plt.imshow(aux)
plt.text(0, 0, "eminen - zuckerberg")

a = fig.add_subplot(1, 2, 2)
plt.axis("off")

aux = abs(zuckerberg - eminem)
plt.imshow(aux)
plt.text(0, 0, "zuckerberg - eminen")
```
